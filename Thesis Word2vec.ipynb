```python
def review_preprocessing(input_review):
    input_review = input_review.astype(str).str.lower() #converting to lower case
    input_review = input_review.astype(str).str.replace('[{}]'.format(string.punctuation), '') #remove punctuation
    input_review = input_review.astype(str).str.replace("[^a-zA-Z#]",' ') #remove special characters
    #input_review = input_review.apply(lambda x: ' '.join([word for word in str(x).split() if not word.isdigit() if not word in stops])) #remove numbers
    input_review = input_review.astype(str).str.strip() #remove spaces
    
    return input_review
```

```python
def remove_stopwords(word_list):
        processed_word_list = []
        for word in word_list:
            for miniword in word:
            #word = word.lower() # in case they arenet all lower cased
                if miniword not in stopwords.words("english"):
                    processed_word_list.append(miniword)
        return processed_word_list
```

```python
df['Data'] =review_preprocessing(df['Data'])
```

```python
dat=[]
data=[]
for i in df['Data']:
    dat.append(i.split(' '))
dat = remove_stopwords(dat)
nested=list()
dat=[dat[i:i+10] for i in range(0, len(dat), 5)]
dat
```

```python
# Data should be in tokenize form.
E1_model = models.Word2Vec (dat, size=150, window=10, min_count=2, workers=10)
E1_model.train(dat,total_examples=len(dat),epochs=10)
```
